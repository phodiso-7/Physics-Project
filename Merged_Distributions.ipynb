## Plots

```python


# Dataframes dictionary for easy iteration
dataframes = {'Signal': sig, 'WW': ww, 'ZZ': zz, 'HZ': hz, 'WWZ': wwz}
colors = {'Signal': 'red', 'WW': 'black', 'ZZ': 'grey', 'HZ': 'blue', 'WWZ': 'green'}

# Plotting variables
variables = ['Jet1_Energy', 'Jet1_Theta', 'Jet1_Phi', 'MissingET.MET', 'Total_InvariantMass']
labels = [r'Energy (GeV)', r'$\theta$ (rad)', r'$\phi$ (rad)', r'Missing $E_T$ (GeV)', r'$M_{4j}$ (GeV)']


# Create subplots for each variable
for variable, label in zip(variables, labels):
    fig, ax = plt.subplots(figsize=(10, 6))
    for name, df in dataframes.items():
        ax.hist(
            df[variable].dropna(),
            bins=50,
            density=True,
            histtype='step',
            linewidth=1.5,
            label=name,
            color=colors[name]
        )
    ax.set_xlabel(label, fontsize=14)
    ax.set_ylabel('', fontsize=14)
    ax.set_title(f'Distribution of {label} for Jet 1', fontsize=16)
    ax.legend(fontsize=12)
    ax.grid(True, alpha=0.3)
    # Customize spines (optional, you can remove spines here if desired)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    plt.show()
```
